```python
#Q4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset
categories = ['Category A', 'Category B', 'Category C', 'Category D']
n_samples_per_category = 100

data = {
    'Category': np.repeat(categories, n_samples_per_category),
    'Value': np.concatenate([
        np.random.normal(loc=50, scale=10, size=n_samples_per_category),
        np.random.normal(loc=60, scale=15, size=n_samples_per_category),
        np.random.normal(loc=55, scale=20, size=n_samples_per_category),
        np.random.normal(loc=65, scale=5, size=n_samples_per_category)
    ])
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='pastel')

plt.title('Distribution of Numerical Variable Across Categories', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset
n_samples = 1000

# Create correlated features
x = np.random.normal(size=n_samples)
y = x * 0.5 + np.random.normal(size=n_samples) * 0.5  # y is correlated with x
z = y * 0.3 + np.random.normal(size=n_samples) * 0.7  # z is correlated with y

# Create a DataFrame
data = {
    'Feature 1': x,
    'Feature 2': y,
    'Feature 3': z
}

df = pd.DataFrame(data)

# Compute the correlation matrix
correlation_matrix = df.corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.show()
```

```python
#Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Display in Jupyter Notebook (remove this line if running as a script)
output_notebook()

# Generate random data
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.rand(n) * 30 + 10  # marker sizes between 10 and 40
colors = ['#%02x%02x%02x' % (int(r), int(g), int(b)) for r, g, b in
          zip(np.random.randint(0, 256, n),
              np.random.randint(0, 256, n),
              np.random.randint(0, 256, n))]

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create the figure
p = figure(title="Random Scatter Plot with Varying Sizes and Colors",
           x_axis_label='X Axis',
           y_axis_label='Y Axis')

# Add circle markers
p.scatter('x', 'y', source=source, size='sizes', color='colors', alpha=0.6)

# Show the plot
show(p)
```

```python
#Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
#fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
#counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

# Display in notebook
output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a figure
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruit', y_axis_label='Count',
           toolbar_location=None, tools="")

# Add vertical bars
p.vbar(x='fruits', top='counts', width=0.6, source=source,
       fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Customize plot
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

```python
#Q4. Create a Bokeh histogram to visualize the distribution of the given data

#data_hist = np.random.randn(1000)
#hist, edges = np.histogram(data_hist, bins=30)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display plot in notebook (remove this line if running as a script)
output_notebook()

# Generate data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create the figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency')

# Add quad glyphs to display histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

```python
#Q5. Create a Bokeh heatmap using the provided dataset.
#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Display in notebook
output_notebook()

# Generate heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays
x_flat = xx.ravel()
y_flat = yy.ravel()
value = data_heatmap.ravel()

# Create ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=value))

# Create figure
p = figure(title="Heatmap Example", tools="", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), match_aspect=True)

# Add rectangles representing heatmap cells
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=linear_cmap('value', Viridis256, low=min(value), high=max(value)),
       line_color=None)

# Show the plot
show(p)
```
